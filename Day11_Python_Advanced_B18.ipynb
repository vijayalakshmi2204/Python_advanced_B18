```python
import numpy as numpy
import pandas as pd
from matplotlib import pyplot as plt
```

```python
simple graph
```

```python
x=[1,2,3]
y=[1,2,3]
plt.plot(x,y)
plt.show()
```

```python
# graph for plotting sales data
#sales vs revenue
#x axis--->sales
#y axis--->revenue
#title:performance evaluation of sales data
```

```python
plt.title('Performance Evaluation Of Sales Data')
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.plot(x,y)
plt.show()
```

```python
adding another axis say z
```

```python
x=[1,2,3]
y=[1,4,9]
z=[10,5,0]

plt.title('Performance Evaluation Of Sales Data')
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
adding name to the plots
```

```python
x=[1,2,3]
y=[1,4,9]
z=[10,5,0]

plt.title('Performance Evaluation Of Sales Data')
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.plot(x,y)
plt.plot(x,z)
plt.legend(['plot of (x,y)','plot of (x,z)'])
plt.show()
```
